```python
import pandas as pd
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:Doodle13!@localhost:5432/Employee_HW')
connection = engine.connect()
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
departments   = pd.read_sql_table('departments',connection)
employees     = pd.read_sql_table('employees',connection)
dept_managers = pd.read_sql_table('dept_manager',connection)
dept_emp      = pd.read_sql_table('dept_emp',connection)
salaries      = pd.read_sql_table('salaries',connection)
titles        = pd.read_sql_table('titles',connection)
```

```python
#histogram of salaries
salaries.salary.plot.hist()
plt.title('Salaries Histogram')
plt.xlabel('Salary')
```

```python
#bar chart salary by title
merge_table = pd.merge(titles, salaries, on="emp_no")
grouped_titles = merge_table.groupby(['title'])
title_by_salary = grouped_titles['salary'].mean()
title_by_salary.plot.bar(title = "Salary by Title")
plt.ylabel("Average Salary")
```
